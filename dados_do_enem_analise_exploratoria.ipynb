Selecionado apenas as colunas que interessam

```
colunas_select = ['NU_INSCRICAO', 'NU_NOTA_MT','NU_NOTA_REDACAO', 'Q001', 'Q002']
```

Criando um novo dataframe para analisar

```
df_selecao = microdados_enem.filter(items=colunas_select)
```

```
df_selecao.head()
```

```
df_selecao = df_selecao.dropna()
```

```
df_selecao.head()
```

Indicadores socioeconômicos 

Q001 - Até que série pai/responsável estudou

Q002 - Até que série a mãe/reponsável estudou

Criando o dicionário

```
dicionario = {'A':	'Nunca estudou', 'B':'Não completou a 4ª série/5º ano do Ensino Fundamental', 
              'C':'Completou a 4ª série/5º ano, mas não completou a 8ª série/9º ano do Ensino Fundamental', 'D':'Completou a 8ª série/9º ano do Ensino Fundamental, mas não completou o Ensino Médio',
              'E':'Completou o Ensino Médio, mas não completou a Faculdade','F':'Completou a Faculdade, mas não completou a Pós-graduação',
              'G':'Completou a Pós-graduação', 'H':	'Não sei'}
```

```
df_selecao['NO_Q001'] = [dicionario[resp] for resp in df_selecao.Q001]
```

```
df_selecao.head(2)
```

```
df_selecao['NO_Q002'] = [dicionario[resp] for resp in df_selecao.Q002]
df_selecao.head(2)
```

Comparando a distribuição do nível de escolaridade dos pais

Utilizando os métodod do DF filter(), groupby(), count(), sort_values()

1- distribuição dos candidatos ENEN 2019 por escolaridade do Pai



```
df_selecao.filter(items= ['NU_INSCRICAO', 'NO_Q001']).groupby('NO_Q001').count().sort_values(by='NU_INSCRICAO', ascending=False)
```

2- distribuição dos candidatos ENEN 2019 por escolaridade do Mãe

```
df_selecao.filter(items= ['NU_INSCRICAO', 'NO_Q002']).groupby('NO_Q002').count().sort_values(by='NU_INSCRICAO', ascending=False)
```

Escolaridade do Pai/Mãe versus nota de matemática

```
df_selecao.filter(items= ['NU_NOTA_MT', 'NO_Q001']).groupby('NO_Q001').mean().sort_values(by='NU_NOTA_MT', ascending=False)
```

```
df_selecao.filter(items= ['NU_NOTA_MT', 'NO_Q002']).groupby('NO_Q002').mean().sort_values(by='NU_NOTA_MT', ascending=False)
```

Quanto maior a escolaridade do pai e mãe maior é a nota de matemática dos candidatos

Fazendo a mesma análise com as notas de redação = NU_NOTA_REDACAO

Do Pai

```
q002redacao = df_selecao.filter(items= ['NU_NOTA_REDACAO', 'NO_Q001']).groupby('NO_Q001').mean().sort_values(by='NU_NOTA_REDACAO', ascending=False)
```

```
q002redacao.plot(title="A escolaridades dos pais impacta positivamente o desempenho do aluno no Enem")
```

Da mãe

```
df_selecao.filter(items= ['NU_NOTA_REDACAO', 'NO_Q002']).groupby('NO_Q002').mean().sort_values(by='NU_NOTA_REDACAO', ascending=False)
```

```
df_selecao.filter(items= ['NU_NOTA_REDACAO', 'NO_Q001']).groupby('NO_Q001').mean().sort_values(by='NU_NOTA_REDACAO', ascending=False).describe()
```
